```python
import cv2.cv2 as cv2
import matplotlib.pyplot as plt
import numpy as np

from dif_aniso import add_noise, dif_aniso, noise_types
```

```python
image = cv2.imread("materiales/T1.png", cv2.COLOR_BGR2GRAY)
image = image.astype(np.float64)

noise_image = add_noise(image=image, noise_type=noise_types.RICIAN_NOISE, intensity=0.3)
fig, ax = plt.subplots(1, 2, figsize=(8.4, 9.6), tight_layout={
        'rect': [0, 0, 1, 0.98]
    }
)
ax[0].imshow(noise_image, cmap='gray')
ax[0].set_title('original')
ax[1].imshow(noise_image, cmap='gray')
ax[1].set_title('ruidosa')
fig.show()
```

```python
denoised_image = dif_aniso(im=noise_image, niter=5, k=40, l=0.25, option=2)
fig, ax = plt.subplots(1, 1, figsize=(8.4, 5.8), tight_layout={
        'rect': [0, 0, 1, 0.98]
    }
)
ax.imshow(denoised_image, cmap='gray')
ax.set_title("denoised")
fig.show()
```
